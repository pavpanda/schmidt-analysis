```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array(["Fire ant", "Sweat bee", "Honey bee", "Yellow jacket wasp", "Bald-faced hornet", "Paper wasp", "Harvester ant", "Tarantula hawk wasp", "Bullet ant"])
y = np.array([1, 1, 2, 2, 2, 3, 3, 4, 4])

plt.style.use('ggplot')
plt.bar(x, y)
plt.xticks(rotation='60')
plt.yticks([1, 2, 3, 4])
plt.ylabel("Schmidt sting pain index")
plt.xlabel("Insect")
plt.title("Schmidt sting pain index by Insect ")
plt.savefig("all.png", bbox_inches="tight")
```
